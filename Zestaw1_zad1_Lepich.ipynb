```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## Zadanie 1
Mając do dyspozycji generator liczb pseudolosowych o rozkładzie jednorodnym $U[0, 1]$

```python
ug = np.random.uniform
```

wygeneruj liczby z rozkładu:
### 1. jednorodnego na przedziale $[a, b]$

```python
def continuous_uniform(a, b):
    def F(x):
        return (b - a)*x + a
    y = ug()
    return F(y)
```

```python
a, b = 0, 10
samples = 50000
bins = 10
X = [continuous_uniform(a, b) for _ in range(samples)]
hist_data = plt.hist(X, bins=bins, density=True)

def f(a, b):
    return 1/(b-a)
    
X = np.linspace(a, b)
plt.plot(X, [f(a, b) for _ in X])
```

### 2. wykładniczego $p(x) = \begin{cases}0, x<0 & \\ \lambda e^{-\lambda x}, x\ge 0 \\ \end{cases} $

```python
def p(lambd):
    def F(x):
        return (1/lambd) * np.log(1/(1-x))
    y = ug()
    return F(y)
```

```python
samples = 50000
bins = 100
lambd = 1

X = [p(lambd) for _ in range(samples)]
hist_data = plt.hist(X, bins=bins, density=True)

def f(x, lambd):
    if (x < 0):
        return 0
    return lambd * (np.e)**(-lambd*x)

X = np.linspace(min(X), max(X))
plt.plot(X, [f(x, lambd) for x in X])
```

### 3. normalnego $N(0,1)$

```python
def N():
    u1, u2 = ug(), ug()
    #y2 = np.sqrt(-2*np.log(u1)) * np.cos(2*np.pi*u2)
    return np.sqrt(-2*np.log(u1)) * np.sin(2*np.pi*u2)
```

```python
samples = 50000
bins = 61

X = [N() for _ in range(samples)]
hist_data = plt.hist(X, bins=bins, density=True)

def f(x):
    return 1/np.sqrt(2*np.pi) * np.exp(-(x**2)/2)

X = np.linspace(min(X), max(X))
plt.plot(X, [f(x) for x in X])
```
