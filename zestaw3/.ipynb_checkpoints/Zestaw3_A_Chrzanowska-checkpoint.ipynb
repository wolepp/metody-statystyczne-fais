## Zadanie 1

```python
import numpy as np
```

```python
def exp(lamb):
    u = np.random.uniform()
    return -np.log(1 - u)/lamb
```

```python
n_iter = 10
signal_time = {}
for i in range(n_iter):
    t = 0
    n_sygnalow = 0
    signal_time[i] = []
    while(t < 90):
        t+=exp(1)
        n_sygnalow+=1
        signal_time[i].append(t)
    
print("Ilość sygnałów: {} \t Suma czasu: {}".format(n_sygnalow, t))
```

```python
from matplotlib import pyplot as plt
```

```python
for i in range(n_iter):
    plt.plot(signal_time[i])
plt.xlabel("czas t∈(0,90)min")
plt.ylabel("liczba wysłanych sygnałów")
plt.title("Liczba wysłanych sygnałów w funkcji czasu")
plt.show()
```

## Zadanie 2

```python
n_tra = 10**4
t_max = [1, 20, 90]
liczba_syg = dict.fromkeys(t_max)

for m in t_max:
    liczba_syg[m] = []
    for i in range(n_tra):
        count = 0
        t = 0
        while(True):
            t += exp(1)
            if(m<t):
                break
            count += 1
        liczba_syg[m].append(count)
    
```

```python
import math

def silnia(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * silnia(n-1)

def poisson(k,t):
    licz = math.pow(t, k)
    mian = silnia(k)
    return licz/mian * math.pow(math.e, (-t))
```

```python
for m in t_max:
    fig, ax = plt.subplots(1,1)
    if max(liczba_syg[m]) == 6:
        bins = 6
    else:
        bins = 10
    ax.hist(liczba_syg[m], density=True, bins=bins, label='Rozkład liczby sygnałów')
    ax.set_title("Rozkład liczby sygnałów dla chwili czasu t = {}".format(m))
    Y = []
    for i in range(max(liczba_syg[m])):
        Y.append(poisson(i,m))
    plt.plot(range(max(liczba_syg[m])), Y, label='Rozkład Poissona')
    plt.show()
```

## Zadanie 3

```python
k = [2, 5, 10, 50]
n_tra = 10**4
czas_oczekiwania = dict.fromkeys(k)

for i in k:
    czas_oczekiwania[i] = []
    for j in range(n_tra):
        count = 0
        t = 0
        while(True):
            count += 1
            if(i == count):
                break
            t += exp(1)
        czas_oczekiwania[i].append(t)
```

```python
for i in k:
    fig, ax = plt.subplots(1,1)
    ax.hist(czas_oczekiwania[i], density=True)
    ax.set_title("Rozkład czasu oczekiwania na k-te zdarzenie k = {}".format(i))
    plt.show()
```
