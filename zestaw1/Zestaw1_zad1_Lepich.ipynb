### 2. wykładniczego $p(x) = \begin{cases}0, x<0 & \\ \lambda e^{-\lambda x}, x\ge 0 \\ \end{cases} $

```python
def p(lambd):
    def F(x):
        return (1/lambd) * np.log(1/(1-x))
    y = ug()
    if (y == 0):
        return 0
    return F(y)
```

```python
samples = 50000
bins = 100
lambd = 1

X = [p(lambd) for _ in range(samples)]
plt.hist(X, bins=bins, density=True, label="dane numeryczne")

def f(x, lambd):
    if (x < 0):
        return 0
    return lambd * (np.e)**(-lambd*x)

X = np.linspace(min(X), max(X))
plt.plot(X, [f(x, lambd) for x in X], label="$\lambda \mathrm{e}^{-\lambda x}$")
plt.title("Rozkład wykładniczy, $\lambda = {}$".format(lambd))
plt.legend(fontsize='large')
plt.savefig("rozklad_wykladniczy.png", facecolor='white', dpi=120)
plt.show()
```

### 3. normalnego $\mathcal{N}(0,1)$

```python
def N():
    u1, u2 = ug(), ug()
    while (u1 == 0):
        u1 = ug()
    #y2 = np.sqrt(-2*np.log(u1)) * np.cos(2*np.pi*u2)
    return np.sqrt(-2*np.log(u1)) * np.sin(2*np.pi*u2)
```

```python
samples = 50000
bins = 61

X = [N() for _ in range(samples)]
plt.hist(X, bins=bins, density=True, label="dane numeryczne")

def f(x):
    return 1/np.sqrt(2*np.pi) * np.exp(-(x**2)/2)

X = np.linspace(min(X), max(X))
plt.plot(X, [f(x) for x in X], label='$\mathcal{N}(0,1)$')
plt.title("Rozkład normalny (0, 1)")
plt.legend()
plt.savefig("rozklad_normalny.png", facecolor='white', dpi=120)
plt.show()
```
